```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures,scale
from sklearn.metrics import mean_squared_error
from pandas.plotting import scatter_matrix

from keras.layers import Dense
from keras.models import Sequential
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('forestfires.csv')
df.drop(['X','Y','month','day','FFMC','DMC','ISI','DC','rain'],1,inplace=True)

#Inputs and Output
X = scale(np.array(df.drop(['area'],1)))
y = np.array(df.area.apply(lambda x: np.log(x+1)))

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = .20,random_state = 42)

df.head()
scatter_matrix(df)
plt.show()
```

```python
#Initializing Neural
model = Sequential()

model.add(Dense(3, input_dim=3, activation='relu'))

model.add(Dense(4, activation='relu'))
model.add(Dense(3, activation='relu'))

model.add(Dense(1, activation='relu'))

model.compile(loss='mean_squared_error', optimizer = 'adam',metrics=['accuracy'])

history = model.fit(X_train,y_train,nb_epoch= 100, verbose=1,batch_size=10)

[test_loss, test_acc] = model.evaluate(X_test, y_test, batch_size=10)

print("For epoch = 1000, and batch size of 10, the accuracy was 0.49 with loss of 3.49")
print("For epoch = 2000, and batch size of 10, the accuracy was 0.0288 with loss of 2.381") 

print("Evaluation result on Test Data : Loss = {}, accuracy = {}".format(test_loss, test_acc))
```
